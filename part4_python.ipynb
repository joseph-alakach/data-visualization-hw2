```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
cancer_data = pd.read_csv("lung_cancer_prediction_dataset.csv")
air_pollution_data = pd.read_csv("global_air_pollution_dataset.csv")
```

```python
# --1
plt.title('PM2.5 AQI Distribution with Density Overlay')
sns.histplot(data=air_pollution_data, x='PM2.5_AQI_Value', bins=30, edgecolor='black', palette='dark',stat="density", label="Histogram")
sns.kdeplot(data=air_pollution_data, x='PM2.5_AQI_Value', color='red', linewidth=2, fill=True, label="Density Plot (KDE)")
plt.xlabel('PM2.5 AQI Value')
plt.legend()
plt.figtext(0, -0.05,
            "This plot represents the distribution of PM2.5 AQI values with density overlay.",
            wrap=False, fontsize=6)
plt.show()
```
