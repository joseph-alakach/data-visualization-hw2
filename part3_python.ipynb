```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
cancer_data = pd.read_csv("lung_cancer_prediction_dataset.csv")
air_pollution_data = pd.read_csv("global_air_pollution_dataset.csv")
```

```python
# --1
plt.title('Lung Cancer Deaths Distribution')
sns.boxplot(y='Annual_Lung_Cancer_Deaths', data=cancer_data)
plt.show();
```

```python
# --2
plt.title('Histogram of PM2.5 AQI Values')
sns.boxplot(y='PM2.5_AQI_Value', data=air_pollution_data)
plt.show();
```

```python
# --3
sns.kdeplot(data=cancer_data, x='Mortality_Rate')
plt.title('Lung Cancer Mortality Rate density')
plt.show()
```
